# Objective

*Background*

In US alone, around 7500 yearly cases of mushrooms poisoning are reported (Source). According to the source, "misidentification of edible mushroom species appears to be the most common cause and may be preventable through education". To avoid expenses for hospitalization and in some cases pointless deaths, you have been hired by US National Health Service to create a machine learning model, that can recognize mushroom types. They want to install this on hand-held devices and to help people make the right choice when mushroom picking.
Concepts to explore

Today, we will put everything we learned in this module and use it to solve a classification problem. The idea of this project is to use transfer learning on an architecture of your choice and fine-tune to predict mushroom types.

You will use this Kaggle dataset https://www.kaggle.com/maysee/mushrooms-classification-common-genuss-images

.


*How to start?*


*Data*

Well, the obvious first steps will be getting the data from Kaggle. There are a number of choices on how to do it, such as downloading images to your machine and then uploading to Drive or using Kaggle API. Once you get your data, start with an EDA, as this will directly feed into design choices for your architecture.
Modeling

My suggestion is that you start with a simple pre-trained architecture, like ResNet18. This will allow you to fine-tune your net faster and if results are not too good, you can try switching to a larger model later. It is recommended that you use PyTorch Lightning or FastAI. Both are equally good for simple problems like this, but PyTorch Lightning will give you more control, better customization ability, and better understanding of your network.

*Requirements:*

- Choose whichever framework you prefer from FastAI, PyTorch Lightning, PyTorch, or Keras.
- As always - EDA
- Use a pre-trained neural net as a backbone of your class
- Train a classifier. Don't forget to fine-tune
- Evaluate inference time
- Visualize results


*Bonus challenges*

- Repeat the process with modifications to your network and see how the results vary.
- Try a different optimizer
- Add an intermediate layer between the backbone and output layer


# additional infos

approaches tested:

- load data with dataset.ImageFolders - avoid creation of custom dataset class.
   - problems: universal transform applied to all images before train/val/test split
   - splitting with random_split etc.: no stratification!
   - doing splits after loading the dataset requires creation of new dataset class for load the train/val/test dataset and apply their dedicated transforms, especially albumentations needs an approach alternative to pytorch native transforms --> new dataset definition becomes wrapper of wrapper of wrapper ... cannot be easily accessed in the dataloader anymore.
   - split with sklearn on indices, load dataset 3 times with transforms for train/val/test, respectively, cut subsets from them with the according indices from sklearn to form the train/val/test dataset. - waste of memory

- splitfolders (copy data to new train/val/test folders), then load each data with dataset.ImageFolders and their corresponding transform.
   - problem: no stratification, extra space on HD


- currently best approach: create df from paths and classes, 2 times sklearn test_train_splits, costom dataset class to prepare for loading albumentation transforms, create datasets by loading images paths defined by the indices collections from the index-split, per dataset-purpose and with their corresponding transform.
